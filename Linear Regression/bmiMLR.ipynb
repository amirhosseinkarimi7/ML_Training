```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

warnings.filterwarnings("ignore")
plt.style.use('ggplot')
```

```python
df = pd.read_csv('InsuranceData.csv')
df.head()
```

```python
df = pd.get_dummies(data= df, drop_first=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['charges'])
y= df['charges']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1455,test_size=0.25)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
model = LinearRegression()
model.fit(X=X_train,y=y_train)
```

```python
model.coef_, model.intercept_
```

```python
y_prediction = model.predict(X_test)
```

```python
r2_score(y_pred=y_prediction,y_true=y_test)
```

```python
mean_absolute_error(y_pred=y_prediction,y_true=y_test), mean_absolute_percentage_error(y_pred=y_prediction,y_true=y_test), mean_squared_error(y_pred=y_prediction,y_true=y_test)
```
